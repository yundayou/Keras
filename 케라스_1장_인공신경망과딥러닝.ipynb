## 1장 인공신경망과 딥러닝
### 개요
#### 인공신경망의 역사
>- 인공지능 : 컴퓨터가 사람처럼 주어진 문제를 해결하기 위해 주어진 데이터를 이용해 스스로 학습하고 판단 할 수 있는 능력을 갖는 것
>- 인공지능을 갖기 위해 사용하는 알고리즘 :  인공신경망(ANN : Artificial Neural Network)

#### 인공신경망과 딥러닝
>- AI vs ML vs 딥러닝
>- 딥러닝 : 인공신경망 알고리즘을 이용하여 문제를 해결하는 방법 중 한 분야
>- 인공신경망 : 머신러닝의 한 분야
>- 머신러닝 :  인공지능의 한분야

##### 딥러닝의 정의
>- 데이터를 학습시켜 모델을 만드는 머신러닝 알고리즘의 집합
>- 비선형 변환으로 구성된 아키텍처를 사용하여 데이터로부터 모델을 만듬
>- **머신러닝**과 **딥러닝**의 차이 : 훈련과정에서 변수의 특징을 추출해서 모델을 만드는 것과 그렇지 않은 것의 차이
>- **머신러닝** : 데이터를 학습시켜 모델을 만들기 전에 변수들에서 특징 추출, 변수들의 특징에 따라 변수 선택, 차원 축소, 주성분 분석, 다중공선성 해결작업
>- **딥러닝** : 학습을 통해 만들어진 모델 사용

### 인공신경망의 구조 
#### 인공신경망에서의 뉴런
##### 뉴런
>- 신경계와 신경조직을 이루는 기본 단위라고 알려진 신경세포
>- 신경계의 모든 작용은 신경세포와 신경세포 간의 상호 작용으로 이루어짐
>- 우리 몸의 내부와 외부에 자극을 가하게 되면 일련의 과정을 통해 뉴런은 자극을 전달, 최종적으로 척수와 뇌 등의 중추신경계에서 처리한 정보를 다시 우리몸으로 전달해 명령을 수행
>- 입력신호와 출력신호는 분명히 다른 신호
>- 학습 : 뉴련과 뉴런 사이의 연결강도가 바뀌는 것

##### 인공뉴런
>- 인간이 누런의 구조에 착안해서 만들어진 인공뉴런도 생물학적 뉴런과 비슷한 구조를 가지고 있음
>- 인공신경망에서 학습에 의하 각 입력에 따른 연결강도가 정해지는데 -> 가중치

##### 활성화 함수
>- 신경계의 뉴런은 들어오는 자극이 일정 크기 이상이 되어야 다음 뉴런으로 전달
>- 활성화 함수 : 선형결합된 값을 다음 출력으로 보내야 할지를 결정
>- 선형결합된 값이 일정 크기를 넘어섰을 경우 다음 출력으로 보내짐

#### 인공신경망의 구조
>- 입력층, 은닉층, 출력층으로 구성되며 각 층의 뉴런들을 **퍼셉트론**이라고 부름
>- 입력츠의 뉴런의 수는 입력 변수의 수, 출력층은 분류문제를 해결할 경우 분류의 수와 일치

##### 얕은 신경망
>- 입력층과 출력층 사이에 숨겨진 계층 :  은닉층
>- 은닉층 :  뉴런과 시냅스로 구성
>- 보통 은닉층의 수가 적거나 없는 신경망 :  얕은 신경망

##### 심층 신경망
>- 입력층과 출력층 사이에 은닉층을 여러개 두어 신경망을 구성한 것
>- 다층신경망이라고도 무름
>- DNN은 딥러닝을 구현하기 위한 알고리즘 중에서 가장 기본이 되는 신ㄱ녕망
>- 학습을 통해서 얻은 모델을 이용하여 결과를 추론하기 까지의 과정을 의미
>- 심층 신경망에서 각 레이어의 뉴런은 이전 레이어의 출력이 가중치와 선형결합되고 활성화 함수를 거쳐 다음 레이어의 입력으로 전달
>- 은닉층이 많으면 가중치를 더 많이 가지므로 예측력이 더 높을 가능성이 크지만, 은닉층이 많다고 해서 무조건 좋다고 말할 수는 없음
>- 은닉층의 수가 너무 많으면 학습하는데 시간이 길어지거나 예측한 값과 실제값이 줄어들지 않는 현상(Gradient Vanishing)등 여러문제점이 발생할 수 있음
>- 인공신경망의 입력으로 전달된 데이터는 은닉층을 통하고, 출력층을 통해 결과를 예측
>- 예측한 값이 실제 값과의 차이를 계산하고 인공신경망은 그 차이를 줄여나가는 방향으로 가중치 들을 조절하면서 학습시키는과정을 거침

#### 구현할 인공신경망 모델의 구조
>- 은닉층 2개와 출력층 하나로 구성
>- 은닉층의 뉴런 50개와 30개
>- 은닉층의 활성화 함수는 시그모이드 함수를 사용, 츌력층의 활성화 함수는 소프트맥스 함수사용
>- 가중치 갱신을 위하여 옵티마이저 Adam을 사용하며, 손실함수는 크로스엔트로피를 사용
#### iris 데이터 불러오기

```python
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
iris_X = iris.data
iris_y = pd.get_dummies(iris.target).to_numpy()

from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(iris_X, iris_y, test_size = 0.3, random_state = 1)
```

#### 싸이킷런을 이용한 구현

```python
mlp.fit(train_X, train_y)
```

```python
mlp.score(test_X, test_y)
```

#### 텐서플로우를 이용한 구현
>- 텐서플로우는 구글에서 만든 딥러닝 라이브러리

```python
! pip install tensorflow
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
x = tf.placeholder(tf.float32, [None,4])
y = tf.placeholder(tf.float32, [None,3])
```

```python
# 은닉층1

W1 = tf.Variable(tf.zeros([4,50]))
b1 = tf.Variable(tf.zeros([50]))
h1 = tf.nn.sigmoid(tf.matmul(x, W1) + b1)

# 은닉층2

W2 = tf.Variable(tf.zeros([50,30]))
b2 = tf.Variable(tf.zeros([30]))
h2 = tf.nn.sigmoid(tf.matmul(h1, W2) + b2)

# 출력층

W3 = tf.Variable(tf.zeros([30,3]))
b3 = tf.Variable(tf.zeros([3]))
h3 = tf.nn.sigmoid(tf.matmul(h2, W3) + b3)
```

```python
cross_entropy = -tf.reduce_sum(y * tf.log(h3), reduction_indices = [1])
loss = tf.reduce_mean(cross_entropy)
train = tf.train.AdamOptimizer().minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

for i in range(1000):
    sess.run(train, feed_dict = {x : train_X, y : train_y})
    if i % 100 == 0:
        tr_loss = sess.run(loss, feed_dict = {x: train_X, y : train_y})
        print(f'Step {i}, {tr_loss:.5f}')
```

```python
correct_prediction = tf.equal(tf.argmax(h3,1), tf.argmax(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print(sess.run(accuracy, feed_dict = {x: test_X, y : test_y}))
```

#### 케라스를 이용한 구현
>- 케라스는 텐서플로우, 씨아노, CNTK 딥러닝 라이브러리 위에서 실행

```python
from tensorflow.keras.models import Sequential
model = Sequential()

from tensorflow.keras.layers import InputLayer, Dense
model.add(InputLayer(4))
model.add(Dense(50, activation = 'sigmoid'))
model.add(Dense(30, activation = 'sigmoid'))
model.add(Dense(3, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_X, train_y, epochs = 1000)
```

```python
model.evaluate(test_X, test_y)
```
